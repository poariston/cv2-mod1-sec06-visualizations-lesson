```python
import pandas as pd
import numpy as np

from sklearn.datasets import load_boston

from pprint import pprint

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
boston = load_boston()
```

```python
data = boston.data
columns = list(boston.feature_names) + ['target']
target = boston.target
```

```python
df = pd.DataFrame(np.column_stack((data, target)), columns=columns)
df.head()
```

```python
def plot_histogram(bins=20, alpha=0.8):
    fig = plt.figure(figsize=(8, 5))
    plt.hist(df['RM'], bins=bins, alpha=alpha)
    plt.title("Room Histogram")
    return fig
    
    

def plot_bar(df, col, figsize=(8, 5), color='purple', alpha=0.8, rotation=75):
    """
    Input
    df: pandas dataframe
    col: name of column you want to plot
    figsize: (tuple) - desired figuresize default = (8, 5)
    ...
    
    Return
    Nothing, but will plot a bar plot
    """
    df[col].value_counts().plot(kind='bar', figsize=figsize, color=color, alpha=alpha)
    plt.xticks(rotation=rotation, fontsize=16)
    plt.yticks(fontsize=16)
    plt.title(col.title().replace("_", " ") + " Bar Chart")
    plt.show()
```

```python
plot_bar(df, 'age_categories')
```

```python
plot_histogram(df, "AGE")
```

### What did you takeaway from what just happened?
- Creating methods and functions can be useful

```python
age_categories = []
for age in df.AGE:
    if age < 40:
        age_categories.append('young')
    elif age < 80:
        age_categories.append('medium')
    else:
        age_categories.append('old')
        
age_categories[:5]
```

```python
df["age_categories"] = age_categories
```

```python
df.head()
```

### Now let's plot a bar chart of age_categories

```python
plot_bar(df, 'age_categories')
```

### Now let's plot a violin plot comparing age to target

[docs here](https://seaborn.pydata.org/generated/seaborn.violinplot.html)

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='target', y='age_categories', data=df)
plt.title("Target by Age Categories")
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
# plot 1 histogram
plt.figure(figsize=(8, 5))
plt.hist(df["RM"], alpha=0.5, bins=20)

# add a mean line
rm_mu = df["RM"].mean()

# plot the mean line (rm_mu)
plt.vlines(rm_mu, 0, 120, colors='r', linestyle='--', linewidth=3)

# let's make a median line
rm_med = df["RM"].median()
plt.vlines(rm_med, 0, 100, colors='green', linestyle='--', linewidth=2)

# let's make a line for quartiles
q1, q3 = np.quantile(df.RM, q=[0.25, 0.75])
plt.vlines(q1, 0, 100, colors='green', linestyle='--', linewidth=2)
plt.vlines(q3, 0, 100, colors='green', linestyle='--', linewidth=2)

# outliers
rm_std = df["RM"].std()

out_right = rm_mu + 3*rm_std
out_left = rm_mu - 3*rm_std

plt.vlines(out_right, 0, 120, colors='r', linestyle='--', linewidth=3)
plt.vlines(out_left, 0, 120, colors='r', linestyle='--', linewidth=3)


plt.show()
```

### Looping through a subplot to populate plots

```python
this, that = plt.subplots(5, 3, figsize=(20, 20))

for index, ax in enumerate(that.flatten()):
    column = df.columns[index]
    ax.hist(df[column], bins=20, label=column)
    ax.set_title(column)
    ax.legend()

plt.show()
```
